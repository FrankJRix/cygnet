```python
import os, cygnet, torch
import torchvision.transforms.v2 as transforms

# paths
train_path = "/home/frx/dataset/image_dataset/train"
vali_path = "/home/frx/dataset/image_dataset/vali"
test_path = "/home/frx/dataset/image_dataset/test"

# folders:
input_path = "input"
mask_path = "mask"
mask_exp2_path = "mask_exp2"
mask_exp4_path = "mask_exp4"

# change this
target_path = mask_exp4_path

input_train_path = os.path.join(train_path, input_path)
target_train_path = os.path.join(train_path, target_path)

input_vali_path = os.path.join(vali_path, input_path)
target_vali_path = os.path.join(vali_path, target_path)

input_test_path = os.path.join(test_path, input_path)
target_test_path = os.path.join(test_path, target_path)

trin = transforms.Compose([transforms.ToImage(), transforms.ToDtype(torch.float32, scale=False)])
trtar = transforms.Compose([transforms.ToImage()])

train_dset = cygnet.CygnoSet(input_train_path, target_train_path, trin, trtar)
print(f"images in train: {len(train_dset)}")
vali_dset = cygnet.CygnoSet(input_vali_path, target_vali_path, trin, trtar)
print(f"images in vali: {len(vali_dset)}")
test_dset = cygnet.CygnoSet(input_test_path, target_test_path, trin, trtar)
print(f"images in test: {len(test_dset)}")

train_dset_transformed = cygnet.CygnoSet(input_train_path, target_train_path)
```

```python
idx = 4199
titles = ["input", "mask"]

train_dset_transformed.compute_stats()

for i, el in enumerate(train_dset_transformed[idx]):
	print(f"=== {titles[i]} stats: ===")
	print(f"shape: {el.shape}")
	print(f"dtype: {el.dtype}")
	print(f"range: {el.min()}, {el.max()}")
	print()
```
